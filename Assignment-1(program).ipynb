```
from sympy import *  #library for symbolic mathematics

#step1 (initialize variables)
x0 = 2 #(x value - random)
epoches = 124
learnrate = 0.01

#step2 (1st derivative of the function)
x=Symbol('x')
y0=x**4+3*x**2+10
y=y0.diff(x)

#calculate value of 1st derivative at x0
for i in range(0,epoches):   # --> step5 (increments the iter i value)
  df=y.subs(x,x0)
  df=round(df,2)

  #step3 (change in x value)
  x1=-1*(learnrate)*df
  x1=round(x1,2)

  #step4 (update x value)
  x0=x0+x1
  x0=round(x0,2)
  
  #step6 (checks to break the loop or not)
  if i>epoches:
    break
  else:
    continue

#step7 (print the x and minimum values)  ***here i didn't get function's min value as '0'***
print('The minimum value obtained is at x =',x0)

# ***
# df=round((y.subs(x,x0)),2)
# print('At x =',x0,', the minimum value of the function',y0,'is',df)
# ***
```
